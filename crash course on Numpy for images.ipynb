### Crash course on NumPy for images

```python
import skimage
from skimage import data

camera = data.camera()
from skimage import io

io.imshow(camera)
```

```python
camera.min(), camera.max(), camera.mean()
```

```python
mask = camera < 80
camera[mask] = 255
```

```python
from skimage import io

io.imshow(camera)
```
